# Extracting the Latitude and Longitude of different locations

```python
#creating a locator agent
locator = Nominatim(user_agent = "myGeocoder")
location = locator.geocode("Nairobi, Kenya")
```

```python
print("Latitude = {}, Longitude = {}".format(location.latitude, location.longitude))
```

```python
# to extract the latitude and longitude of all 161 regions;
# all region names should be in a single variable i.e through transposing the df
city = city.T
city.head()
```

```python
#changing item names to be column names of the dataset
city.rename(columns = city.iloc[0], inplace = True)
city.drop(city.index[0], inplace = True)
city.head()
```

```python
#region names are indexes. Changing region names to seperate columns;
city = city.reset_index()

#renaming index to column location
city = city.rename(columns = {'index': 'Location'})
city.head()
```

```python
city.columns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#nomatim extracts longitude and latitude one at a time
from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(locator.geocode, min_delay_seconds = 1)
```

```python
#creating a location column
city['location'] = city['Location'].apply(geocode)

#create a lat, long and altitude from location column - returns tuple
city['point'] = city['location'].apply(lambda loc: tuple(loc.point)if loc else None)

#split point column into lat, long and altitude columns
city[['latitude', 'longitude', 'altitude',]] = pd.DataFrame(city['point'].tolist(), index = city.index)

city.head()
```

```python
#droping unnecessary columns
city = city.drop(['location', 'point', 'altitude'], axis = 1)
city.columns
```
